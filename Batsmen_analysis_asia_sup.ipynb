```python
ax=df[df['Strike Rate']<100.00].sort_values(by='Strike Rate', ascending=False)['Player Name'].head(10)
```

```python
a = df.groupby(['Country'],as_index=False)['Matches'].sum().sort_values(by='Matches',ascending=False)
```

```python
ax=sns.barplot(x='Country',y='Matches',data=a,palette='rocket')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Matches Played by Countries")
plt.show()
```

```python
a = df.sort_values(by='Played',ascending=False).head(5)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(x='Player Name',y='Played',data=a,palette='rocket')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player with Most Matches Played")
plt.show()
```

```python
a = df.sort_values(by='Highest Score',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Highest Score',data=a,palette='icefire_r')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player by Highest Score")
plt.show()
```

```python
a = df.sort_values(by='Strike Rate',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Strike Rate',data=a,palette='husl')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player by Highest Score")
plt.show()
```

```python
a = df.sort_values(by='Centuries',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Centuries',data=a,palette='flare')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player with most Centuries")
plt.show()
```

```python
a= df.groupby(['Country'], as_index=False)['Highest Score'].max().sort_values(by='Highest Score', ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Country',y='Highest Score',data=a,palette='viridis_r')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Highest Score from every Country")
plt.show()
```
