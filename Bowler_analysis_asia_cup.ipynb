```python
a = df.groupby(['Country'],as_index=False)['Matches'].sum().sort_values(by='Matches',ascending=False)
ax=sns.barplot(x='Country',y='Matches',data=a,palette='rocket')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Total Matches Played By Bowlers of Each Countries")
plt.show()
```

```python
a = df.sort_values(by='Played',ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(x='Player Name',y='Played',data=a,palette='rocket')
plt.title("Top 5 Bowler with Most Matches Played")
for bars in ax.containers:
    ax.bar_label(bars)
#plt.title("Top 5 Bowler with Most Matches Played")
plt.show()
```

```python
a = df.sort_values(by='Wickets',ascending=False).head(5)
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.barplot(x='Player Name',y='Wickets',data=a,palette='viridis_r')
plt.title("Top 5 Bowler with Most Wickets")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
a = df.sort_values(by='Overs',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Overs',data=a,palette='flare')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Players with most over Bowled")
plt.show()
```

```python
a = df.sort_values(by='Bowling Average',ascending=True).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Bowling Average',data=a,palette='pastel')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player with best Bowling Average")
plt.show()
```

```python
a = df.sort_values(by='Strike Rate',ascending=True).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Player Name',y='Strike Rate',data=a,palette='pastel')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 5 Player with Good Strike Rate")
plt.show()
```

```python
a= df.groupby(['Country'], as_index=False)['Wickets'].max().sort_values(by='Wickets', ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
ax=sns.barplot(x='Country',y='Wickets',data=a,palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Highest Wickets Taker from every Country")
plt.show()
```
